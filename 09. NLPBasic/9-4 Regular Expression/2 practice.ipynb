```python
import re
```

```python
# . 임의의 문자 표현 a.c이면 aac,abc,alc ...etc
r = re.compile('a.c')
r.search('kkk')
```

```python
r.search('abc')
```

```python
# ? 앞의 문자가 있거나 없거나.. ab?c 이면 abc,ac
r = re.compile('ab?c')
r.search('abbc')
```

```python
r.search('ac')
```

```python
# * 앞의 문자가 0개이상일 경우
# ab*c -> ac abc abbc abbbbbbc... b의 값이 무한할수 도 있음
r = re.compile('ab*c')
r.search('a')
```

```python
r.search('ac')
```

```python
r.search('abc')
```

```python
r.search('abbbbbbbbbc')
```

```python
# + *과 유사. 대신 최소 1개는 있어야함.
# ab+c -> ac는 안됨. abc abbbbbc 가능
r = re.compile('ab+c')
r.search('abc')
```

```python
r.search('ac')
```

```python
r.search('abbbbbbbc')
```

```python
# ^ 시작 문자 지정. ^ab이면 ab로 시작하는 경우 매치
r = re.compile('^ab')

r.search('bbc')
r.search('zab')
```

```python
r.search('abz')  
```

```python
# {숫자,숫자} ab{2,8}c이면 a와 c 사이에 b가 2~8인 문자열 매치
r = re.compile('ab{2,8}c')
r.search('ac')
r.search('abc')
r.search('abbbbbbbbbc')
```

```python
r.search('abbbbbbbc')
```

```python
#{숫자,} 숫자 이상 반복만 매치 a{2,}bc이면 bc전 a가 2개 이상일때
r = re.compile('a{2,}bc')
r.search('bc')
r.search('aa')
```

```python
r.search('aabc')
```

```python
#[] 안의 문자중 한개와 매치. [abc]이면 a,b,c 하나 혹은 전부...
# [a-zA-z] 알파벳 전부, [0-9]는 숫자 전부
r = re.compile('[abc]')
r.search('zzz')
```

```python
r.search('a')
```

```python
r.search('aaaaaaaaa')
```

```python
r.search('bacc')
```

```python
r = re.compile('[a-z]')
r.search("AAAA")
r.search('111')
```

```python
r.search('aBC')
```

```python
#[^문자] ^기호 뒤에 붙은 문자들을 제외한 모든 문자를 매치하는 역할.
# [^abc] a,b,c를 제외
r = re.compile('[^abc]')

r.search('a')
r.search('ac')
r.search('c')
```

```python
r.search('aq')
```

```python
#re,match() re.search() 차이 시작점이 맞아야 match에 걸림
r = re.compile('ab.')
r.match('kkkabc')
```

```python
r.match('abckkk')
```

```python
text = """이름 : 김출서
번호 : 010 - 1234 -5678
성별 : 남"""

re.findall('\d+',text) # 숫자만 리턴
```

```python
# sub는 정규 표현식 패턴과 일치하는 문자열을 찾아 다른 문자열로 교체.
# 정제 과정에 많ㅇ ㅣ사용, 특수문자가 섞여있는경우 제거 하고싶을때  사용..
text = "Regular expression : A regular expression, regex or regexp[1] (sometimes called a rational expression)[2][3] is, in theoretical computer science and formal language theory, a sequence of characters that define a search pattern."

preprocessed_text = re.sub('[^a-zA-Z]', ' ', text)
print(preprocessed_text)
```
